```python
import numpy as np
#Define a function to calculate the promotion
def calculate_promotion(dataset):
    #Calculate the promotion based on the input features
    dataset['promotion'] = np.where((dataset['previous_year_rating'] > 2) & 
                                    (dataset['length_of_service'] > 2) & 
                                    (dataset['awards_won'] > 0), 
                                    'Yes', 
                                    'No')
    return dataset
```
